```python
import re
```

```python
genes = open('my_genes.txt','r')
hcontigs = open("hcontigs.txt", 'w')
gnome = open("spades_raw.filter10cov.fasta")
genes_str = genes.read()
genes_array = re.split("//", genes_str)[:-1]
data = {}
```

```python
for element in genes_array:
    if re.search("CDS", element): 
        contig = re.split("\;", element)
        contig_id = re.findall('"([^"]+)', contig[2])[0]
        line = '>' + contig_id + "\n"
        data[line] = None
        hcontigs.write(line)

hcontigs.close()
```

```python
def concatenate_list_data(list):
    result = ''
    for element in list:
        z = re.match("(>\w+)\W", element)
        if z is None: 
            result += str(element) 
        else: 
            result += '\n' + str(element) + '\n'
    return result 

with open("spades_raw.filter10cov.fasta", "r") as file:
    alist = file.read().split()

fasta = concatenate_list_data(alist).strip()
parts = fasta.split()

    
hcontigs_data = open("hcontigs-data.txt", 'w')

for i in range(0, len(parts), 2):
    print(i)
    print(i+1)
    data[parts[i]] = parts[i+1]
    
for k in data:
    print('header' + k)
    if data[k] is not None:
        hcontigs_data.write(k+'\n')
        hcontigs_data.write(data[k])
    
hcontigs_data.close()

```
