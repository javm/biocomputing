```python
output = []

def process_line(f):
    line = f.readline()
    parts = line.strip().split("\t")
    # print(parts)
    return {'line': line, 'parts': parts}
with open("RAL-304.view", 'r') as f:
    r = process_line(f)
    line = r['line']
    parts = r['parts']
    while(line):
        if( int(parts[4]) >= 100 and int(parts[5]) >= 10 and float(parts[6]) <= 2):
            output.append(parts[0]+"\t"+parts[1]+"\t"+parts[4]+"\t"+parts[5]+"\t"+parts[6])
        r = process_line(f)
        line = r['line']
        parts = r['parts']
        
print(output)

output_ral = open('output-ral.view', 'w')
for i in range(0, len(output)):
    output_ral.write(output[i]+'\n')  
output_ral.close()

print(i)
print(len(output))

    
    
```
