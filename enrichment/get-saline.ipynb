```python
import csv
import re
import sys
```

```python
csv.field_size_limit(sys.maxsize)
results = {}
salinos = []
with open('./data/meta_salino.txt') as salinos_file:
    salinos = salinos_file.readlines()
    salinos = [s.strip('\n') for s in salinos]
print(salinos)
```

```python
with open('./data/test.csv', newline='') as test:
    testreader = csv.reader(test)
    next(testreader)
    row_num = 0
    
    for row in testreader:    
        matches = []
        headers = re.split(">", row[4])
        headers.pop(0)
        #print(headers)
        #headers_num = 0;
        for h in headers:
            species = re.split("\|", h)
            species = species[1].strip('[]')
            matches.append(species)
        results[row[0]] = {'matches': matches, 'hits': row[1]}
```

```python
for k in results:
    r = results[k]
    r['salinos'] = 0
    for o in r['matches']:
        if(o in salinos):
            r['salinos'] += 1

print(results)
```

```python
target = dict()
for k in results:
    match = results[k]
    if(int(match['hits']) == int(match['salinos'])):
        target[k] = match
print(target)
```

```python
output = open('./data/salinos_matches.txt', 'w')
for k in target:
    m_s = ''
    for m in target[k]['matches']:
        m_s = m_s +' '+m
    output.write(k +'\t'+ target[k]['hits'] +'\t'+m_s+'\n')
output.close()
```
